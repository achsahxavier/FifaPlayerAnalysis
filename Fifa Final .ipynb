```python
def clean_weight(x):
    
    if isinstance(x,str):
        return(x.replace('lbs', ''))
    return(x)

fifa['Weight'] = fifa['Weight'].apply(clean_weight).astype('float')
fifa['Weight'].fillna((fifa['Weight'].mean()), inplace = True) 
```

```python
fifa['Weight'].head()
```

```python
# defining a function for cleaning the wage column
def conv(num):
    if num[-1]=='M':
        return float(num[1:-1])*1000000
        
    elif num[-1]=='K':
        return float(num[1:-1])*1000
    else:
        return float(num[1:])


fifa['Value'] = fifa['Value'].apply(lambda x: conv(x))
fifa['Wage'] = fifa['Wage'].apply(lambda x: conv(x))
```
